```python
#COLOR regognition
import cv2
cap=cv2.VideoCapture(0)
width=cap.get(3)
hight=cap.get(4)
FONT=cv2.FONT_HERSHEY_COMPLEX

while True:
    ret,frame=cap.read()
    if ret == True:
        frame=cv2.addWeighted(frame,.5,frame,0,10)
        HSV=cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
        CX,CY=int(width/2),int(hight/2)
        H=HSV[CX,CY,0]
        color=None
        if H<5:
            color='Red'
        elif H<22:
            color='ORANGE'
        elif H<33:
            color='YELLOW'
        elif H<78:
            color='GREEN'
        elif H<131:
            color='Blue'
        elif H<170:
            color='VIOLET'
        else:
            color='RED'
        
        B,G,R=frame[CX,CY,:]
        frame=cv2.circle(frame,(CX,CY),5,(int(B),int(G),int(R)),3)
        frame=cv2.putText(frame,color,(300,30),FONT,1,(int(B),int(G),int(R)),2)
    

            
        cv2.imshow('image',frame)
        if cv2.waitKey(1)==ord('q'):
            break
    else: 
        break
cap.release()
cv2.destroyAllWindows()
```
